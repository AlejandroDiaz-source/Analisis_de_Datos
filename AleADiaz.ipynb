## Definir grafico

```python

def make_graph(data, title):
    # Extraer las fechas y los precios de cierre de las acciones
    fechas = data.index
    precios_cierre = data['Close']

    # Crear el gráfico
    plt.figure(figsize=(10, 6))
    plt.plot(fechas, precios_cierre)
    plt.title(title)
    plt.xlabel('Fecha')
    plt.ylabel('Precio de cierre')

    # Mostrar el gráfico
    plt.show()

# Obtener los datos de las acciones de Tesla
tesla_data = yf.download('TSLA')

# Llamar a la función make_graph para graficar los datos de acciones de Tesla
make_graph(tesla_data, 'Datos de acciones de Tesla')
```

```python
# Obtener los datos de las acciones de gme
gme_data = yf.download('GME')

# Llamar a la función make_graph para graficar los datos de acciones de gme
make_graph(gme_data, 'Datos de acciones de GameStop')
```
